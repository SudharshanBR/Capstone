### 2. Recreating Descriptive Statistics

#### Calculating Mean Percentage Error

- Why? Well, MPE uses actual values instead of absolutes. This may result in positive and negative values offsetting each other.
- That will be a clear indication of the presence of bias in the forecasts.
- Plus, there were no 0 values present in the entire data and so the disadvantage of MPE will not be experienced.

```python
 dataframe['Percent Error'] = ((dataframe['Customer Orders'] - dataframe['Forecasted Demand'])/dataframe['Customer Orders']) * 100.0
```

#### Histogram of Percent Errors

```python
fig1DF = dataframe.copy(deep=True) #Fig1 has its own DF.

plotName = "Histogram Of Percent Errors"


if plt.fignum_exists(plotName):
    plt.close(plotName)
    
fig,ax = plt.subplots(num=plotName,figsize=(8,4))


x = np.linspace(-1600,250,200)
ax.hist(fig1DF['Percent Error'].values, bins=67, range=[-1600,200],density=False,  edgecolor='Black');
ax.yaxis.tick_right()
ax.set_xlabel('Percent Error')
ax.set_ylabel('Counts')
ax.set_title('Histogram of Percent Errors')
ax.yaxis.set_label_position("right")
```

### 3. Intel's perspective and their approaches


#### a) Double Booking


- Pick a Product Offering, preferably E to stay consistent with the paper.
- Drop unrelated columns.
- Scatter + Line plot the customer orders.
- The Intuition : Customer inflates the forecast for the same SKU but for 2+ generations.

```python
fig3DF = dataframe.loc[dataframe['Product Offering'] == 'E'] #Fig3 has its own DF.
fig3DF = fig3DF.drop(['Distribution Center','Product Offering','SKU','ASP Group','Percent Error'],axis=1)

fig3DF['Gen2'] = fig3DF.loc[fig3DF['Generation'] == 2,'Customer Orders']
fig3DF['Gen3'] = fig3DF.loc[fig3DF['Generation'] == 3,'Customer Orders']
fig3DF['Gen4'] = fig3DF.loc[fig3DF['Generation'] == 4,'Customer Orders']
fig3DF['Gen5'] = fig3DF.loc[fig3DF['Generation'] == 5,'Customer Orders']
fig3DF['Forecast - Demand'] = fig3DF['Forecasted Demand'] - fig3DF['Customer Orders']

# fig3DF
```

```python
from scipy.linalg import pinv

k_max = 9 # a0, a1, a2,b1, b2
X = len(gen2DF.index)
X_mat = np.empty((N,2*k_max+1),dtype =float)
X_mat[:,0] = 1
for k in range(1,k_max+1):
    X_mat[:,2*k-1] = np.cos(2*np.pi*k*x/X)
    X_mat[:,2*k] = np.sin(2*np.pi*k*x/X)    
a = pinv(X_mat)@y
yhat = X_mat@a
ax.plot(x,yhat,label=f"k={k}")
ax.legend()
```
